```python
import numpy as np #backend de pandas y cálculos mátematicos.
import pandas as pd #pandas para el manejo de datasets
pd.set_option('display.max.columns', 100) #configuración de pandas para mostrar todas las cols (100 en este caso)
#%matplotlib inline #configuración del notebook para dibujar dentro del notebook
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importo el Dataset al Notebook

```python
data = pd.read_csv('C:/Users/lucas/OneDrive/Documentos/Diplomatura/Dataset/adult.data',names = ["age","workclass","fnlwgt","education","education-num","marital-status","occupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country","salary"])
data.head()
```

# EJERCICIOS A RESOLVER:

## 1 - ¿Cuantos hombres y mujeres están representados en el dataset (sex feature)?


```python
#Contamos cantidades de cada sexo.
data['sex'].value_counts()
```

```python
#Realizamos un grafico tipo "PIE" representando los resultados e imprimimos la Respuesta.
data['sex'].value_counts().plot.pie()
print('Rta: Hay %i hombres y %i mujeres en el Dataset.' % (data['sex'].value_counts()[0],data['sex'].value_counts()[1]))
```

# --------------------------------------------------------------------------------------------------

## 2 - ¿Cuál es el promedio de edad de las mujeres? ¿Y de los hombres?

```python
#Calculo los promedios en conjunto.
data.groupby("sex")["age"].mean()
```

```python
#Graficación.
data.groupby("sex")["age"].mean().plot.bar()

#Calculo los promedios por separado.
promedio_mujeres = round(data[data['sex'] == ' Female']['age'].mean(), 2)
promedio_hombres = round(data[data['sex'] == ' Male']['age'].mean(), 2)
print('Rta: El promedio de edad de las Mujeres es de %s años; el de los hombres es %s años.' % (promedio_mujeres, promedio_hombres))
```

# --------------------------------------------------------------------------------------------------

## 3 - ¿Cuál es el porcentaje de ciudadanos Alemanes? (native-country feature)?

```python
#Cantidad de Alemanes.
alemanes = data[data["native-country"] == " Germany"]
cantidad_de_alemanes = alemanes["native-country"].count()

#Cantidad de Personas en total.
poblacion_total = data["native-country"].count()

#Calculo Porcentaje de Alemanes respecto a Población Total.
porcentaje_alemanes = round(cantidad_de_alemanes * 100 / poblacion_total, 2)
print("Rta: Existe un %s%% de ciudadanos alemanes." % (porcentaje_alemanes))  

#Creando Gráfico.
nacionalidades = ['Ciudadanos de otros Paises', 'Ciudadanos Alemanes']
cantidades = [poblacion_total, cantidad_de_alemanes]
explode = [0, 0.4]
plt.pie(cantidades, labels=nacionalidades, explode=explode, autopct='%1.1f%%', shadow=True, startangle=0)
plt.show()
```

# --------------------------------------------------------------------------------------------------

##  4-5 ] - Media y Desviación Estándar de la edad para los que ganan más de 50K y menos.

```python
#Cantidad, Promedio y Desviación Estándar de los que ganan más de 50K.
ganan_mas = data[data['salary'] == ' >50K']
media_ganan_mas = ganan_mas['age'].mean()
desviacion_estandar_mas = ganan_mas['age'].std()

#Cantidad, Promedio y Desviación Estándar de los que ganan menos de 50K.
ganan_menos = data[data['salary'] == ' <=50K']
media_ganan_menos = ganan_menos['age'].mean()
desviacion_estandar_menos = ganan_menos['age'].std()

#Impresión de Resultados.
print("""Para los que ganan más de 50K:
        MEDIA: %i años
        DESVIACIÓN ESTANDAR: %i""" % (media_ganan_mas, desviacion_estandar_mas))
print('--------------------------------')
print("""Para los que ganan menos de 50K:
        MEDIA: %i años
        DESVIACIÓN ESTANDAR: %i""" % (media_ganan_menos, desviacion_estandar_menos))
```

## 6 - ¿Las personan que ganan más de 50K, tienen al menos educación secundaria?

```python
#Agrupamos según salario y educación.
estadistica_ganancia_estudio = data.groupby('salary') ['education'].value_counts()

#Impresión.
print('Rta: No necesariamente tienen que tener Educación Secundaria para ganar mas de 50K los individuos.')
print('--------------------------------------------------------------------------------------------------')
estadistica_ganancia_estudio
```

## 7 - Mostrar estadísticas de edad por cada raza y por cada género.  Amer-Indian-Eskimo.

```python
#Creando Grupos según Filtros solicitados.
grupos_varios = data.groupby(['sex', 'race'])
estadisticas_edad_grupos = grupos_varios['age'].describe()

#Definiendo por pasos la caracteristica requerida del Individuo.
amer_indian_eskimo = data[data['race'] == ' Amer-Indian-Eskimo']
amer_indian_eskimo_male = amer_indian_eskimo[amer_indian_eskimo['sex'] == ' Male']
amer_indian_eskimo_male_max = amer_indian_eskimo_male['age'].max()

#Imprimiendo Resultados.
print('Rta: La maxima edad de los hombres de raza \"Amer-Indian-Eskimo\" es de %i años.' % (amer_indian_eskimo_male_max))
print('------------------------------------------------------------------------------')
print('\"Estadísticas de Edad, según el Género y la Raza\".')
estadisticas_edad_grupos
```

## 8 - ¿De la gente que gana más de 50K, hay más casados o solteros?

```python
#Filtramos por las Ganancias primero.
ganan_mas = data[data["salary"]==" >50K"]

#Filtramos los Casados.
civil_casados = ganan_mas[ganan_mas["marital-status"].str.startswith(" Married")]
civil_casados_cantidad = civil_casados["marital-status"].count()

#Filtramos los Solteros.
civil_solteros = ganan_mas[(ganan_mas["marital-status"].str.contains(" Never-married")) | (ganan_mas["marital-status"].str.contains('Divorced'))]
civil_solteros_cantidad = civil_solteros['marital-status'].count()

#Imprimimos la Evaluación.
if civil_casados_cantidad > civil_solteros_cantidad:
    print('Rta: La cantidad de personas Casadas, es mayor que la de Solteros.')
else:
    print('Rta: La cantidad de personas Solteras, es mayor que la de Casados.')
    
#Imprimimos Resultados.
print('------------------------------------------------------------------')
print('CANTIDAD SOLTEROS: %i' % (civil_solteros_cantidad))
print('CANTIDAD CASADOS: %i' % (civil_casados_cantidad))

#Creando el Gráfico.
clases = ['Solteros', 'Casados']
proporcion = [civil_solteros_cantidad, civil_casados_cantidad]
explode = [0.4, 0]
plt.pie(proporcion, labels= clases, explode = explode, autopct ='%1.1f%%', shadow = True, startangle = 0)
plt.show()
```

## 10 - Tiempo de trabajo promedio para los que ganan poco y mucho, por país. Y en Japón?

```python
#Filtrando Grupo en función del Salario y la Nacionalidad.
pd.set_option('display.max_rows', 100)
grupo_segun_salario_nacionalidad = data.groupby(['salary', 'native-country'])
estadistica_tiempo_trabajo = grupo_segun_salario_nacionalidad['hours-per-week'].describe()

#Mostrando Resultados.
estadistica_tiempo_trabajo
```

```python
#Calculando Estadísticas de Japón.
japan_promedio_mayor = estadistica_tiempo_trabajo.loc[' >50K', ' Japan']['mean']
japan_promedio_menor = estadistica_tiempo_trabajo.loc[' <=50K', ' Japan']['mean']

#Impresión Estadísticas de Japón.
print("""Rta: En Japón tenemos:
        -Sueldo Mayor a 50K: un Promedio de %i horas trabajadas Diarias.
        -Sueldo Menor o Igual a 50K: un Promedio de %i horas trabajadas Diarias.
        """ % (japan_promedio_mayor, japan_promedio_menor))

#Graficación.
clases = ['Mayor Sueldo', 'Menor Sueldo']
valores = [japan_promedio_mayor, japan_promedio_menor]
plt.bar(range(2), valores, edgecolor='black')
plt.xticks(range(2), clases, rotation=60)
plt.title("Horas de Trabajo entre Mayor Sueldo VS Menor Sueldo")
plt.show()
```
